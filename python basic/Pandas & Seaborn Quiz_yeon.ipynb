#### 문제2. 데이터 다루기

```python
Olympic.head()
```

```python
# 2.1 Year별 Data수를 확인하시오.

Olympic.groupby('Year').size()
```

```python
# 2.2 NOC가 KOR인 선수들 별 출전 횟수를 출력하시오.

kor = Olympic[Olympic['NOC']=='KOR'].groupby(Olympic['Name'])
kor.sum()
```

```python
# 2.3 Year가 2012이고 NOC가 KOR이며 Medal은 Gold인 선수들의 이름,나이,성별,키,몸무게,메달 을 출력하시오.
kor_g = Olympic[(Olympic['Year']==2012)&(Olympic['NOC']=='KOR')&(Olympic['Medal']=='Gold')]
kor_g = kor_g.drop(['Team','NOC','Year'], axis=1)
kor_g = kor_g[['Name','Age','Sex','Height','Weight','Medal']]
kor_g.head()
```

```python
kor_g.describe()
```

```python
# 2.4 Medal이 Gold, Silver이며 NOC가 KOR,USA,CHN인 데이터만 head()로 출력하시오.

med = Olympic[(Olympic['NOC'].isin(['KOR','USA','CHN']))&(Olympic['Medal'].isin(['Gold','Silver']))]
med.head()
```
